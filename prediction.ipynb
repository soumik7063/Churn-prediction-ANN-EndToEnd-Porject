```python
import pandas as pd
import tensorflow as tf
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import pickle
```

```python
# load model
model = load_model('model.h5')

# load encoder and scaller

with open ('label_encoder_gender.pkl','rb') as file:
    labelabel_encoder_gender = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    'Geography':"France",
    'Gender':"Male",
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000,
}
```

```python
with open('one_hot_encoder.pkl','rb') as file:
    one_hot_encoder = pickle.load(file)
```

```python
encoded_geo = one_hot_encoder.transform([[input_data['Geography']]]).toarray()
```

```python
geo_encoded_df = pd.DataFrame(encoded_geo,columns=one_hot_encoder.get_feature_names_out())
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender']= labelabel_encoder_gender.transform([input_df['Gender']])
```

```python
final_input_df = pd.concat([input_df.drop(['Geography'],axis=1),geo_encoded_df],axis=1)
```

```python
scaled = scaler.transform(final_input_df)
scaled
```

```python
# predict
prediction = model.predict(scaled)
```

```python
prediction_prob = prediction[0][0]
prediction_prob
```

```python
if prediction_prob >0.5 :
    print("customer is likely to churn")
else :
    print("Customer will not churn")
```
